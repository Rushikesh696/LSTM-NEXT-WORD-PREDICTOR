### Import required libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import pickle
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

### Initialize Tokenizer
- Create a new Keras Tokenizer to convert text into sequences.
- Build the vocabulary index by fitting the tokenizer on the sample text.

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([sample])
```

##### Check the legnth of the words

```python
len(tokenizer.word_index)
```

```python
# for sentence in sample.split('\n'):
  # print(sentence)
```

### Create Input Sequences
- Generate n-gram sequences for each sentence to prepare training data.

```python
input_sequence = []
for sentence in sample.split('\n'):
  tokenize_sentence = tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenize_sentence)):
    n_gram = tokenize_sentence[:i+1]
    input_sequence.append(n_gram)
```

```python
input_sequence
```

### Find Maximum Sequence Length
Calculate the length of the longest sequence to use for padding.

```python
max_len = max([len(i) for i in input_sequence])
padded_input_sequences = pad_sequences(input_sequence, maxlen=max_len, padding = 'pre')
```

```python
padded_input_sequences
```

### Prepare Features (X)
Remove the last word from each sequence to use as input features for training.

```python
x = padded_input_sequences[:,:-1]
x
```

### Prepare Labels (y)
Extract the last word from each sequence to use as the target label for training.

```python
y = padded_input_sequences[:,-1]
y
```

```python
x.shape
```

```python
y.shape
```

### One-Hot Encode Labels (y)
Convert target labels into one-hot encoded vectors for multi-class classification.

```python
y= to_categorical(y, num_classes=len(tokenizer.word_index)+1)
y
```

```python
y.shape
```

### Build LSTM Model
Create a sequential model with an embedding layer, LSTM layer, and dense output layer.

```python
model = Sequential()
model.add(Embedding(len(tokenizer.word_index)+1, 100))  # removed input_length
model.add(LSTM(150))
model.add(Dense(len(tokenizer.word_index)+1, activation='softmax'))
```

### Compile the Model
Configure the LSTM model for training by specifying the loss function, optimizer, and evaluation metrics.

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Build Model & View Summary
Build the model with the specified input shape and display its architecture.

```python
model.build(input_shape=(None, 56))
model.summary()
```

```python
x
```

### Train the Model
Fit the LSTM model on the prepared input and output data for multiple epochs.

```python
model.fit(x,y, epochs = 200)
```
